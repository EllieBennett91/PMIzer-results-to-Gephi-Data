```python
import pandas as pd
PmiResults = pd.read_csv('NA_Checked_Lemma_NPMI2_10.csv', sep='\t')
PmiResults
```

```python
#make a df of only word 1
s1 = PmiResults[['word1']]
s1
```

```python
#make a df of only word2
s2 = PmiResults[['word2']]
s2
```

```python
#rename s2 columne to 'word1'
s2.columns = ['word1']
s2
```

```python
Nodes = pd.concat([s1, s2], ignore_index=True)
Nodes
```

```python
#appends the original 'word2' column below the 'word1' column
#Nodes = s1.append(s2, ignore_index=True)
#Nodes
```

```python
#remove duplicates
Nodes = Nodes.drop_duplicates(subset=['word1'], keep='last')
Nodes
```

```python
#re-order the index to start from 1
Nodes = Nodes.reset_index(drop=True)
Nodes
```

```python
#add ID column (the row index plus one)
#Nodes['Label'] = Nodes.index + 1
Nodes['Id'] = Nodes.index + 1
Nodes
```

```python
#rename word1 colum as 'label'
Nodes.columns = ['Label','Id']
Nodes
```

```python
#reorder columns (Id, then Label)
Nodes = Nodes[['Id', 'Label']]
Nodes
```

```python
#remove duplicates
#Nodes = Nodes.drop_duplicates(subset=['Label'], keep='last')
#Nodes
```

```python
#save as a .csv file
savefile =  'NodeListCheckedLemmaNADN.csv'
with open('output/' + savefile, 'w', encoding="utf-8") as w:
    Nodes.to_csv(w, index=False)
```
